```python
%matplotlib inline
import numpy as np
import pymc3 as pm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t
#plt.style.use('seaborn-darkgrid')
print('Running on PyMC3 v{}'.format(pm.__version__))
```

## Data

Create a dataset with t-distributed random data. <br>
We create a test and a control set, like in a situation where we are testing the performance of changes to our catalogs/website.<br>
The synthetic data rapresents the tot_dmd_amt that customer spent, and we label it with the group each customer belongs to.<br>

Scipy.stats and Padas are used to sample from the t-distribution and store the data in a dataframe.

```python
#control = np.random.normal(loc=65, scale=10, size=10000)
#test = np.random.normal(loc=69, scale=14, size=10000)
control = t.rvs(df=30, loc=65, scale=10, size=10000)
test = t.rvs(df=30, loc=69, scale=12, size=10000)

vals = np.r_[control,test]
grp = np.r_[['control'] * len(control), ['test'] * len(test)] 

data = pd.DataFrame({'tot_dmd_amt':vals, 'group':grp})

data.sample(10)
```

Let's see the data in a distribution plot

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,10))
ax = sns.distplot(data.loc[data.group=='test', 'tot_dmd_amt'] , label='test')
ax = sns.distplot(data.loc[data.group=='control', 'tot_dmd_amt'], label='control' )
plt.legend()
plt.show()
```

### Equation for the student distribution:
<br>

Student-t distribution specified by:
1. mean  
2. precision (inverse-variance)  
3. degrees-of-freedom parameter  
    * small  => heavy tail
    * large  => normal distribution

<h1><center>$f(x|\mu,\lambda,\nu) = \frac{\Gamma(\frac{\nu + 1}{2})}{\Gamma(\frac{\nu}{2})} \left(\frac{\lambda}{\pi\nu}\right)^{\frac{1}{2}} \left[1+\frac{\lambda(x-\mu)^2}{\nu}\right]^{-\frac{\nu+1}{2}}$</center></h1>

With PyMc3 we can build a Bayesian model by specifing how the likelihood and priors should be parametrised.<br>

Here we specify how the mean and std_dev of the t-distribution should be parametrised.<br>
Following Krushke we use a normal dist for the mean and a uniform prior for the std_dev.

```python
mu_m = data.tot_dmd_amt.mean()
mu_s = data.tot_dmd_amt.std() * 2
sigma_low = 1
sigma_high = 20

with pm.Model() as model:
    # declare the mean for the two groups
    test_mean = pm.Normal('test_mean', mu_m, sd=mu_s)
    control_mean = pm.Normal('control_mean', mu_m, sd=mu_s)
    # declare the std_dev for the two groups
    test_std = pm.Uniform('test_std', lower=sigma_low, upper=sigma_high)
    control_std = pm.Uniform('control_std', lower=sigma_low, upper=sigma_high)
```

The deg-of-freedom parameter is parametrized by an exponential. The author suggests starting with a parameter that guarantees an heavy tail.<br>
**Note** As usual different stats packages use different definitions for the distribution parameters - lambda vs rate - one the inverse of the other.<br>
**Note 2** data is synthetic so let't start far away from the real value and see if we can get it to converge close to the real value.

```python
pm.kdeplot(np.random.exponential(50, size=10000), shade=0.5);
```

```python
with model:
    v = pm.Exponential('v', 1/50.0)
```

*From PyMC3 Docs:*
... parameterizes the Student-T in terms of precision, rather than standard deviation, we must transform the standard deviations before specifying our likelihoods.

Lets declare the distribution for our target variable. Student-T.

```python
with model:
    lambd_1 = test_std**-2
    lambd_2 = control_std**-2

    test_dist = pm.StudentT('test', nu=v, mu=test_mean, lam=lambd_1, 
                            observed=data.loc[data.group=='test', 'tot_dmd_amt'])
    control_dist = pm.StudentT('control', nu=v, mu=control_mean, lam=lambd_2, 
                               observed=data.loc[data.group=='control','tot_dmd_amt'])
```

Now we can calculate the difference of meand, std_devs and the effect size. This are deterministic in nature but will inherit the distribution properties.

```python
with model:
    diff_of_means = pm.Deterministic('difference of means', test_mean - control_mean)
    diff_of_stds = pm.Deterministic('difference of stds', test_std - control_std)
    effect_size = pm.Deterministic('effect size',
                                   diff_of_means / np.sqrt((test_std**2 + control_std**2) / 2))
```

```python
with model:
    trace = pm.sample(5000, cores=4)
```

### Plots to analyse the results

#### Investigate trace plots and results for the parameters

```python
pm.traceplot(trace);
```

#### Plot posterior distributions for the model parameters

```python
pm.plot_posterior(trace, varnames=['test_mean','control_mean', 'test_std', 'control_std', 'v'],
                  color='#87ceeb');
```

#### Plot posteriors for the deterministic values and compare to the null of 0 difference.

```python
pm.plot_posterior(trace, varnames=['difference of means','difference of stds', 'effect size'],
                  ref_val=0,
                  color='#87ceeb');
```

#### Credible values for the parameters

```python
pm.forestplot(trace, varnames=['test_mean',
                               'control_mean']);
```

```python
pm.forestplot(trace, varnames=['test_std',
                               'control_std',
                               '']);
```

#### Print the summary. AKA what the boss cares about.

```python
pm.summary(trace,varnames=['difference of means', 'difference of stds', 'effect size'])
```

```python
%matplotlib inline
import numpy as np
import pymc3 as pm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t
#plt.style.use('seaborn-darkgrid')
print('Running on PyMC3 v{}'.format(pm.__version__))
```

## Data

```python
#control = np.random.normal(loc=65, scale=10, size=10000)
#test = np.random.normal(loc=69, scale=14, size=10000)
control = t.rvs(df=30, loc=65, scale=10, size=10000)
test = t.rvs(df=30, loc=69, scale=12, size=10000)

vals = np.r_[control,test]
grp = np.r_[['control'] * len(control), ['test'] * len(test)] 

data = pd.DataFrame({'tot_dmd_amt':vals, 'group':grp})

data.sample(10)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,10))
ax = sns.distplot(data.loc[data.group=='test', 'tot_dmd_amt'] , label='test')
ax = sns.distplot(data.loc[data.group=='control', 'tot_dmd_amt'], label='control' )
plt.legend()
plt.show()
```

### Equation for the student distribution:
<br>

Student-t distribution specified by:
1. mean μ 
2. precision (inverse-variance) λ 
3. degrees-of-freedom parameter ν 
    * small ν => heavy tail
    * large ν => normal distribution

<h1><center>$f(x|\mu,\lambda,\nu) = \frac{\Gamma(\frac{\nu + 1}{2})}{\Gamma(\frac{\nu}{2})} \left(\frac{\lambda}{\pi\nu}\right)^{\frac{1}{2}} \left[1+\frac{\lambda(x-\mu)^2}{\nu}\right]^{-\frac{\nu+1}{2}}$</center></h1>

```python
mu_m = data.tot_dmd_amt.mean()
mu_s = data.tot_dmd_amt.std() * 2
sigma_low = 1
sigma_high = 20

with pm.Model() as model:
    # declare the mean for the two groups
    test_mean = pm.Normal('test_mean', mu_m, sd=mu_s)
    control_mean = pm.Normal('control_mean', mu_m, sd=mu_s)
    # declare the std_dev for the two groups
    test_std = pm.Uniform('test_std', lower=sigma_low, upper=sigma_high)
    control_std = pm.Uniform('control_std', lower=sigma_low, upper=sigma_high)
```

```python
pm.kdeplot(np.random.exponential(30, size=10000), shade=0.5);
```

```python
with model:
    v = pm.Exponential('v', 1/29.0)
```

```python
with model:
    lambd_1 = test_std**-2
    lambd_2 = control_std**-2

    test_dist = pm.StudentT('test', nu=v, mu=test_mean, lam=lambd_1, 
                            observed=data.loc[data.group=='test', 'tot_dmd_amt'])
    control_dist = pm.StudentT('control', nu=v, mu=control_mean, lam=lambd_2, 
                               observed=data.loc[data.group=='control','tot_dmd_amt'])
```

```python
with model:
    diff_of_means = pm.Deterministic('difference of means', test_mean - control_mean)
    diff_of_stds = pm.Deterministic('difference of stds', test_std - control_std)
    effect_size = pm.Deterministic('effect size',
                                   diff_of_means / np.sqrt((test_std**2 + control_std**2) / 2))
```

```python
with model:
    trace = pm.sample(5000, cores=4)
```

```python
pm.traceplot(trace);
```

```python
pm.plot_posterior(trace, varnames=['test_mean','control_mean', 'test_std', 'control_std', 'v'],
                  color='#87ceeb');
```
